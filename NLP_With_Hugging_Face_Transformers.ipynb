```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")

text = "Vaksinasi COVID-19 sangat penting untuk peningkatan imunitas masyarakat."

candidate_labels = ["kesehatan", "politik", "pendidikan", "ekonomi"]

result = classifier(text, candidate_labels=candidate_labels, multi_label=True)

for label, score in zip(result['labels'], result['scores']):
    print(f"{label}: {score:.4f}")
```

```python
from transformers import pipeline

generator = pipeline("text-generation", model='gpt2')

prompt = "We will learning about Data Science"

generated_text = generator(prompt, max_length=50, num_return_sequences=3)

print(generated_text[0]['generated_text'])
```

```python
from transformers import pipeline

unmasker = pipeline("fill-mask", model='bert-base-uncased')

prompt = "This course will teach you all about [MASK] models and their applications in real-world scenarios."

# Mengisi kata yang hilang
results = unmasker(prompt, top_k=3)

# Menampilkan hasil
print("Possible words to fill in [MASK] are:")
for result in results:
    print(f"{result['sequence']} (Score: {result['score']:.4f})")
```

```python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)

text = "Nama saya Hardianto Tandi Seno dan sekarang berstatus pengangguran"

result = ner(text)

for entity in result:
    print(f"Entity: {entity['word']}, Label: {entity['entity_group']}")
```

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")

context = "bernafas"

question = "Manusia memanfaatkan paru-paru untuk?"

result = qa_pipeline(question=question, context=context)
print(f"Jawaban: {result['answer']}")
```

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")

classifier("Jokowi melakukan blunder yang fatal menjelang pergantian presiden")
```

```python
from transformers import pipeline

summarizer = pipeline("summarization")
summarizer(
"""
    A balanced diet and regular exercise are two major factors that can improve quality of life.
    Many studies have shown that people who eat nutritious foods and exercise regularly have a lower risk
    of developing chronic diseases such as diabetes, hypertension, and heart disease.
"""
)
```

```python
from transformers import pipeline

translator = pipeline("translation_id_to_en", model="Helsinki-NLP/opus-mt-id-en")

text_to_translate = "Aku tidak punya uang 100 ribu"
result = translator(text_to_translate)

print(result[0]['translation_text'])
```
